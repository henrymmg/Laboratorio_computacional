```python
% pylab inline
from mpl_toolkits.mplot3d import Axes3D
```

```python
radio=2
N=1000
x=[]
y=[]
z=[]
r=[]
xx=[]
yy=[]
zz=[]

for i in range(N):
    x.append(0)
    y.append(0)
    z.append(0)
    r.append(0)
    xx.append(0)
    yy.append(0)
    zz.append(0)
```

```python
for i in range(N):
    
    x[i]=np.random.uniform(-radio,radio)
    y[i]=np.random.uniform(-radio,radio)
    z[i]=np.random.uniform(-radio,radio)
    r[i]=np.sqrt((x[i]*x[i])+(y[i]*y[i])+(z[i]*z[i]))
    
    if(r[i]<radio):
    
        xx[i]=x[i]
        yy[i]=y[i]
        zz[i]=z[i]
        
        
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xx, yy, zz)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')       
```
